```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://data.wprdc.org/datastore/dump/9c817094-ab87-4ef8-b92a-e0e9289e7553")
district_names = {
    1: "Central Business District",
    2: "Strip District",
    3: "Shadyside",
    4: "Squirrel Hill",
    5: "Oakland",
    6: "South Side Flats",
    7: "Bloomfield",
    8: "Highland Park",
    9: "East Liberty",
    10: "North Shore",
    11: "Lawrenceville",
    12: "Greenfield",
    14: "Hazelwood"}
df["District Name"] = df["district"].map(district_names)
df = df.sort_values("District Name")
plt.figure(figsize=(12, 6))
for district in df["District Name"].unique():
    district_data = df[df["District Name"] == district]
    plt.bar(district_data.index, district_data["acres"], label=district)
plt.xlabel("X-axis")
plt.ylabel("Size (Acres)")
plt.title("Size of Individual EMS Facilities by District in Pittsburgh")
plt.legend(title="Districts", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Each Town gets ranked based on number of Facilities and Total size of the facilities per town.
#Both totals get added together to figure out just based off of EMS facilities which town in Pittsburgh has the best Public Safety
df = pd.read_csv("https://data.wprdc.org/datastore/dump/9c817094-ab87-4ef8-b92a-e0e9289e7553")
district_names = {
    1: "Central Business District",
    2: "Strip District",
    3: "Shadyside",
    4: "Squirrel Hill",
    5: "Oakland",
    6: "South Side Flats",
    7: "Bloomfield",
    8: "Highland Park",
    9: "East Liberty",
    10: "North Shore",
    11: "Lawrenceville",
    12: "Greenfield",
    14: "Hazelwood"}
df["District Name"] = df["district"].map(district_names)
facility_counts = df.groupby("District Name").size().reset_index(name="Facility Count")
facility_counts["Facility Points"] = facility_counts["Facility Count"] * 10
def calculate_size_points(size):
    if size < 1000:
        return 5
    elif 1000 <= size < 2000:
        return 10
    elif 2000 <= size < 3000:
        return 15
    elif 3000 <= size < 4000:
        return 20
    elif 4000 <= size < 5000:
        return 25
    elif 5000 <= size < 6000:
        return 30
    else: 
        return 35
df["Size Points"] = df["acres"].apply(calculate_size_points)
size_points = df.groupby("District Name")["Size Points"].sum().reset_index()
combined_points = pd.merge(facility_counts, size_points, on="District Name")
combined_points["Total Points"] = combined_points["Facility Points"] + combined_points["Size Points"]
combined_points = combined_points.sort_values("Total Points", ascending=False)
print(combined_points)
plt.figure(figsize=(12, 6))
plt.bar(combined_points["District Name"], combined_points["Total Points"])
plt.xlabel("District Name")
plt.ylabel("Total Points")
plt.title("Total Points by District in Pittsburgh (Facility and Size Combined)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
